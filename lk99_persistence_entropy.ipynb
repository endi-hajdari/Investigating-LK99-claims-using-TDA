## Persistence Entropy Using `ripser` and `persim`: LK-99

In this section of code, we visualize the **persistent homology** of the LK-99 atomic structure and compute the **persistent entropy** from the diagrams using the normalized point cloud dataset. Moreover, the code below does the following:
- Loads 3D atomic coordinates from `normalized_point_cloud.txt`; 
- Computes persistence diagrams using `ripser` (up to dimension 2); 
- Visualizes the resulting topological features using `persim`; and
- Calculates the persistent entropy from persistence diagrams to get a quantitative value representing structural order/disorder.

```python
# Importing all relevant libraries required for all blocks of code
from ripser import ripser as rp
from persim import plot_diagrams       
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the LK-99 point cloud
X = np.loadtxt("normalized_point_cloud.txt")

# Computing the persistent homology up to H2
result = rp(X, maxdim=2)
diagrams = result['dgms']

# Plotting the persistence diagrams
plot_diagrams(diagrams, show=False)
plt.title("Persistence Diagram of LK-99 Atomic Structure (H₀–H₂)")
plt.xlabel("Birth")
plt.ylabel("Death")
plt.grid(True)
plt.show()

# Defining the function to calculate persistent entropy for persistence diagram
def persistent_entropy(diagram):
    diagram = diagram[np.isfinite(diagram[:, 1])]   # Filtering out features with infinite death times as they are often the H0 features
    lifespans = diagram[:, 1] - diagram[:, 0]       # Calculating the lifespan of each topological feature
    tot = np.sum(lifespans)                       # If all features are zero-length, return 0 entropy
    if tot == 0:
        return 0
    probs = lifespans / tot                       # Normalizing the lifespans into probabilities
    return -np.sum(probs * np.log(probs))           # Computing the entropy

for dim, dgm in enumerate(diagrams):                # For loop to print the persistent entropy for each dimension
    E = persistent_entropy(dgm)
    print(f"Persistent Entropy (H{dim}): {E:.4f}")
```

### Comparing the Persistent Entropies of Known Superconductors to LK-99

```python
# Defining the point cloud files and their labels
files = {
    "LK-99": "normalized_point_cloud.txt",
    "Perovskite Cuprate": "1001421_point_cloud.txt",
    "Thallium Cuprate": "1001420_point_cloud.txt",
    "Bismuth Cuprate": "1001409_point_cloud.txt"
}

# Persistent entropy calculator
def persistent_entropy(diagram):
    diagram = diagram[np.isfinite(diagram[:, 1])]
    lifespans = diagram[:, 1] - diagram[:, 0]
    total = np.sum(lifespans)
    if total == 0:
        return 0
    probs = lifespans / total
    return -np.sum(probs * np.log(probs))

# Computing each material's entropies
entropy_summary = {}
for name, filename in files.items():
    X = np.loadtxt(filename)
    result = rp(X, maxdim=2)
    diagrams = result["dgms"]
    entropy_summary[name] = [persistent_entropy(dgm) for dgm in diagrams]

# Extracting each material's H1 entropies
labels = list(entropy_summary.keys())
entropies_H1 = [v[1] for v in entropy_summary.values()]  # H1 entropies

# Plotting a bar chart to better visualize comparison
plt.figure(figsize=(10, 5))
plt.bar(labels, entropies_H1, color="mediumseagreen")
plt.title("Persistence Entropy (H₁) Comparison: LK-99 vs. Known Superconductors")
plt.ylabel("Entropy")
plt.xticks(rotation=15)
plt.grid(True)
plt.tight_layout()
plt.show()
```
